# 03 - Filtro Blur

```python
from __future__ import print_function, division
from builtins import range

import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Cargamos la famosa imagen de Lena
img = mpimg.imread('lena.png')
plt.imshow(img)
plt.show()
```

```python
# la pasamos a grises
bw = img.mean(axis=2)
plt.imshow(bw, cmap='gray')
plt.show()
```

```python
# Creamos el filtro Gaussiano
W = np.zeros((20, 20))
for i in range(20):
    for j in range(20):
        dist = (i - 9.5)**2 + (j - 9.5)**2
        W[i, j] = np.exp(-dist / 50.)
```

```python
# Veamos el aspecto del filtro
plt.imshow(W, cmap='gray')
plt.show()
```

```python
# Hacemos la convolucion 2D
out = convolve2d(bw, W)
plt.imshow(out, cmap='gray')
plt.show()
```

```python
# vemos que la imagen queda 'borrosa' o 'difuminada'
# ademas hay un marco negro alrededor
print(out.shape)
# despues de la convolucion la imagen de salida es  N1 + N2 - 1
```

```python
# podemos hacer que sea del mismo tamaño con mode='same'
out = convolve2d(bw, W, mode='same')
plt.imshow(out, cmap='gray')
plt.show()
print(out.shape)
```

```python
# en color sería
out3 = np.zeros(img.shape)
for i in range(3):
    out3[:,:,i] = convolve2d(img[:,:,i], W, mode='same')
plt.imshow(out3)
plt.show() 
```
