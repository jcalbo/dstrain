# Ejemplo de t-SNE (al problema de XOR)

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
```

```python
# Funcion que genera datos distribuidos segun XOR
def get_xor_data():
    X1 = np.random.random((100, 2))
    X2 = np.random.random((100, 2)) - np.array([1, 1])
    X3 = np.random.random((100, 2)) - np.array([1, 0])
    X4 = np.random.random((100, 2)) - np.array([0, 1])
    X = np.vstack((X1, X2, X3, X4))
    Y = np.array([0]*200 + [1]*200)
    return X, Y
```

```python
# Vemos es aspecto de los datos y comprobamos que no son facilmente separables
X, Y = get_xor_data()

plt.scatter(X[:,0], X[:,1], s=100, c=Y, alpha=0.5)
plt.show()
```

```python
tsne = TSNE(perplexity=50)
Z = tsne.fit_transform(X)
plt.scatter(Z[:,0], Z[:,1], s=100, c=Y, alpha=0.5)
plt.show()
```

Ejecutandolo varias veces vemos que la separacin es mas sencilla (aunque no del todo).
