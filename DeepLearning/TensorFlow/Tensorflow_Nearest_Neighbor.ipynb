# Nearest Neighbor - Ejemplo

Basado en: https://github.com/aymericdamien/TensorFlow-Examples/

```python
import numpy as np
import tensorflow as tf

# Cargamos los datos de MNIST usando la clase "input_data"
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/home/jorge/data/MNIST/", one_hot=True)
```

```python
# Usamos solo una parte de los datos de MNIST
Xtr, Ytr = mnist.train.next_batch(5000) #5000 para training 
Xte, Yte = mnist.test.next_batch(200) #200 para testing
```

```python
# Definimos las entradas al grafo
xtr = tf.placeholder("float", [None, 784])
xte = tf.placeholder("float", [784])
```

```python
# Para el calculo de Nearest Neighbor usamos distancia L1 

distancia = tf.reduce_sum(tf.abs(tf.add(xtr, tf.negative(xte))), reduction_indices=1)
```

```python
# Prediccion: Obtenemos el indice con menor distancia (es decir el vecino mas cercano - nearest neighbor)
prediccion = tf.arg_min(distancia, 0)

accuracy = 0.

# Inicializamos las variables en tensorflow
init = tf.global_variables_initializer()
```

```python
# Entrenamos

with tf.Session() as sess:
    sess.run(init)

    # loop sobre los datos de test
    for i in range(len(Xte)):
        # Obtenemos el nearest neighbor
        nn_index = sess.run(prediccion, feed_dict={xtr: Xtr, xte: Xte[i, :]})
        # Obtenemos la etiqueta de clase (label) del nearest neighbor y lo comparamos con el valor verdadero
        print "Test", i, "Prediccion:", np.argmax(Ytr[nn_index]), \
            "Clase real:", np.argmax(Yte[i])
        # Calculamos la accuracy
        if np.argmax(Ytr[nn_index]) == np.argmax(Yte[i]):
            accuracy += 1./len(Xte)
    print "Completado!"
    print "Accuracy:", accuracy
```
