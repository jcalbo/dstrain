## Ejemplo para activar los controles de imagen en matplotlib y ajuste de tamaño

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook 
fig_size = plt.rcParams["figure.figsize"]
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
fig_size[0] = 6
fig_size[1] = 4
plt.rcParams["figure.figsize"] = fig_size
plt.show()
```
